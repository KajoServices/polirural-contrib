# Using Google translate to translate topics

```python
from googletrans import Translator
translator = Translator()
translated = translator.translate('business account', dest='sk')
#print(translated.text)
```

```python
#f = open("/home/dzon/kajo/topic_list_slim-model_26396topics_topn10_dep5_thr0.5.txt", "r")
f = open("/home/dzon/kajo/semantic-explorer/data/topic_list_v3.txt", "r")
i = 0
topics_en = []
topics_sk = []
for line in f:
    i += 1
    #if i > 5:
    #    break
    line = line.strip()
    line = line.split("_")
    line = " ".join(line)
    topics_en.append(line)
    translated = translator.translate(line, src='en', dest='sk')
    topics_sk.append(translated.text)
    print(line, " --- ", translated.text)
```

```python
import pandas as pd

df = pd.DataFrame()
df["topics_en"] = topics_en
df["topics_sk"] = topics_sk
```
