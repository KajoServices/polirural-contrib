```python
for lang in ['cs', 'el', 'en', 'es', 'fi', 'it', 'lv', 'nl', 'pl', 'sk']:
    print("\n\n\n==================================\n==================================\nLANGUAGE: ", lang)
    start = time.time()
    ta = TextAnalyzer(lang=lang)
    print('Finished creating instance %.4f s' % ((time.time()-start)))
    print("number of topics: ", len(ta.TOPICS_LIST))
    for dirpath, dirnames, filenames in os.walk("/home/dzon/kajo/" + lang +"-texts"):
        i = 0
        for filename in [f for f in filenames if f.endswith(".txt")]:
            text_file = os.path.join(dirpath, filename)
            print("\n------------\n")
            print (text_file)
            # load some text
            with open(text_file, encoding="utf-8", mode="r") as fp:
                text = fp.read() 
            print("LEN OF TEXT: ", len(text))
          
            keywords = 15
            topics_extracted = 10
            extracted_keywords = ta.get_keywords(text, keywords=keywords, raw_keywords=True)
            start = time.time()
            extracted_keywords_lemma = ta.get_keywords(text, keywords=keywords)
            print("Extracting keywords: ",time.time()-start)    
            print("\nKEYWORDS: ", len(extracted_keywords))
            for (t,d) in sorted(extracted_keywords.items(), key=lambda kv: kv[1],reverse=True):
                print(t,d)
            start = time.time()
            extracted_topics_mp = ta.get_topics_wmd_mp(text, keywords=keywords, lemma=False, keywords_dic=extracted_keywords_lemma)
            print("\nExtracting topics MP: ",time.time()-start)    
            start = time.time()
            extracted_topics = ta.get_topics_wmd(text, keywords=keywords, lemma=False, keywords_dic=extracted_keywords_lemma)
            print("Extracting topics: ",time.time()-start) 
            print()
            
            print("Topics MP ----- topics")
            for j,(tp_mp, tp) in enumerate(zip(extracted_topics_mp, extracted_topics)):
                print(j, tp_mp, " ---- ", tp)
                
            #print(ta.topics(text))    
```
