# check manually generated topics list - if the topics (terms) exist in the model

```python
from gensim.models import Word2Vec
model=Word2Vec.load("/media/dzon/Data/Jiri-models/wv-sg1hs0.bin")
```

```python
# topic list manually generated from 
filepath = 'topic_list.txt'
# del topics
topics = []
with open(filepath) as fp:
    lines = fp.readlines()
for line in lines:    
    line = line.strip()
    line = line.lower()
    if line != '':
        
        tmp_list = line.split(',')
        for tmp_topic in tmp_list:

            tmp_topic = tmp_topic.strip()
            if tmp_topic == '':
                continue
            try: 
                similar = model.wv.similar_by_vector(tmp_topic, topn=1)
                topics.append(tmp_topic)
                print(tmp_topic)
            except :
                similar = [(None,0)]
                print('!!!' + tmp_topic)
with open('topic_list_checked.txt', 'w') as f:
    for item in topics:
        f.write("%s\n" % item)            
```
