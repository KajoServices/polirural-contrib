# make lemma word embedding smodel
* in `reduce fasttext SK model` notebook we filtered fasttext SK model and left there only slovak words (according to found dictionaries of existing slovak words)
* for every word in the model, we get lemma (using udpipe) - we process only first 500_000 words from the model (most important)
* for every lemma which was produced by at least 2 words we take these 2 words and average their vector which will be the new vector for the lemma word
* the final model is again sorted (the position of the lemma was defined as the min position from the words which transfer to the lemma) and contains 83099 SK words (lemmas)

```python
from gensim.models import Word2Vec, KeyedVectors
import time
import numpy as np
```

```python
model = KeyedVectors.load_word2vec_format('/mnt/data/data/models-fasttext/cc.sk.300-slim.vec.bin', binary=True)
```

```python
words = model.index2word
len(words)
words[0:200]
```

```python
# lemma test
import spacy_udpipe
nlp = spacy_udpipe.load("sk")

for i in range(200):
    doc = nlp(words[i])

    print(words[i], doc[0].lemma_.lower())
```

```python
# process first 500_000 words in fasttext SK model
from tqdm import tqdm, tqdm_notebook
lemmas = {}
word_lemma = {}
for word in tqdm(words[:500000]):
    doc = nlp(word)
    lemma = doc[0].lemma_.lower()
    word_lemma[word] = lemma
    if lemma in lemmas:
        lemmas[lemma] += 1
    else:
        lemmas[lemma] = 1
len(lemmas)        
```

```python
import operator
sorted_lemmas = sorted(lemmas.items(), key=operator.itemgetter(1), reverse=True)
```

```python
number = 83080
count = 20
sorted_lemmas[number:number+count]
```

```python
lemma_words = {}
for word in word_lemma.keys():
    if word_lemma[word] in lemma_words:
        lemma_words[word_lemma[word]].append(word)
    else:
        lemma_words[word_lemma[word]]=[word]
            
```

```python
len(lemmas)
lemma_words['mať']
```

```python
 model.vocab['nemali'].index
```

```python
# compute average vectors from words with same lemma - include only lemmas with at least 2 words
tmp_words = [None] * 500000
tmp_vecs = [None] * 500000
for lemma in lemma_words.keys():
    lemma_words_list = lemma_words[lemma]
    if len(lemma_words_list)<2:
        continue
    tmp_vectors = []
    tmp_indexes = []
    for word in lemma_words_list:
        tmp_vectors.append(model[word])  
        tmp_indexes.append(model.vocab[word].index)
    tmp_words[min(tmp_indexes)] = lemma
    tmp_vecs[min(tmp_indexes)] = list(np.average(tmp_vectors, axis=0))
      
```

```python
len(tmp_vecs[1])
```

```python
#final_words = list(filter(None, tmp_words))
final_words = [i for i in tmp_words if i] 
#final_vecs = list(filter(None, tmp_vecs.any()))
final_vecs = [i for i in tmp_vecs if i != None] 
print(len(final_vecs), len(final_words))
len(final_vecs[0])
```

```python
import gensim
new_vector_space = gensim.models.keyedvectors.Word2VecKeyedVectors(300)
new_vector_space.add(final_words,final_vecs)
```

```python
slim_model_name = '/mnt/data/data/models-fasttext/cc.sk.300-lemma_min2_500000.vec.bin'
new_vector_space.save_word2vec_format(slim_model_name, binary=True)
```
