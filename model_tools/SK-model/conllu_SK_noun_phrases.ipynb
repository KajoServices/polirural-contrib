# Manual tagging of (Slovak) noun_chunks  - preparation for creating of the syntax_iterators.py

```python
# go through 
from io import open
from conllu import parse_incr
import spacy_udpipe, spacy, pytextrank

#nlp = spacy_udpipe.load("sk")
#spacy_udpipe.download("sk") # download English mode



data_file = open("/home/dzon/kajo/spacy_multilang/UD_Slovak-SNK/sk_snk-ud-train.conllu", "r", encoding="utf-8")
for tokenlist in parse_incr(data_file):
    print()
    print(tokenlist.metadata["text"])
    text = tokenlist.metadata["text"]
    #doc = nlp(tokenlist.metadata["text"])
    #for token in doc:
    #    print(token.text, token.lemma_, token.pos_, token.dep_, token.tag_)
    sentence = ""
    for token in tokenlist:
        sentence += str(token["id"]) + "-" + token["form"] + " "
    print(sentence) 
    text += ";" + sentence
    noun_chunk_ids = input("noun chunks (IDs separated by space; noun chunks separated by colon , - e.g. '2 3,7 8 9') :")
    if noun_chunk_ids == "":
        continue
    nc_list = noun_chunk_ids.split(",")
    for i,nc in enumerate(nc_list):
        print("noun chunk #",i+1)
        text += ";nc#" + str(i+1) + " " + nc
        nc_ids = nc.strip().split(" ")
        for nc_id in nc_ids:
            print(" --->id: {} {} POS: {} DEP: {}".format(tokenlist[int(nc_id)-1]["form"], tokenlist[int(nc_id)-1]["id"], tokenlist[int(nc_id)-1]["upostag"], tokenlist[int(nc_id)-1]["deps"]))
            text += "-{},{},{},{}".format(tokenlist[int(nc_id)-1]["form"], tokenlist[int(nc_id)-1]["id"], tokenlist[int(nc_id)-1]["upostag"], tokenlist[int(nc_id)-1]["deps"])
    with open('/home/dzon/kajo/spacy_multilang/UD_Slovak-SNK/sk_snk-ud-train-noun_chunks.csv', 'at', encoding='utf8') as f:        
        f.write(text + "\n")    
```
