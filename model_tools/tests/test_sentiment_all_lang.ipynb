```python
for lang in ['cs', 'el', 'en', 'es', 'fi', 'it', 'lv', 'nl', 'pl', 'sk']:
    print("\n\n\n==================================\n==================================\nLANGUAGE: ", lang)

    tp = TextProcessor(lang=lang)

    for dirpath, dirnames, filenames in os.walk("/home/dzon/kajo/" + lang +"-texts"):
        i = 0
        for filename in [f for f in filenames if f.endswith(".txt")]:
            text_file = os.path.join(dirpath, filename)
            print("\n------------\n")
            print (text_file)
            # load some text
            with open(text_file, encoding="utf-8", mode="r") as fp:
                text = fp.read() 
            sentiment_score = tp.polarity(text, level='t', full=True)
            #print("SENTIMENT: ", sentiment_score)
            print("SENTIMENT: ")
            for sent in sentiment_score:
                print(sent)
```
